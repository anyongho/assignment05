## import packages for plotting graphs and manipulate data
### And write matplotlib incline for plot error

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as ndimage
%matplotlib inline

from scipy import signal
from skimage import io, color
from skimage import exposure
```

## read file image and make a convolution kernel

```python
file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)

ker = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
im_conv = signal.convolve2d(im_gray, ker, boundary='symm', mode='same') 
```

## plot the 4 images for basic image

```python
p1 = plt.subplot(2,2,1)
p1.set_title('color image')
plt.imshow(im_color)
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('gray image')
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
p3.set_title('convolution kernel')
plt.imshow(ker, cmap='gray')
plt.axis('off')

p4 = plt.subplot(2,2,4)
p4.set_title('convolution result')
plt.imshow(im_conv, cmap='gray')
plt.axis('off')

plt.show()
```

## plot the Smoothing image make gaussian_filter kernel

```python
plt.imshow(im_color, interpolation='nearest')
plt.axis('off')
plt.title('Original image')
plt.show()

img = ndimage.gaussian_filter(im_color, sigma=(5,5,0), order=0)
plt.imshow(img, interpolation='nearest')
plt.axis('off')
plt.title('Smoothing image')
plt.show()
```

## make rotate kernel(my own) + x,y gradient using sobel

```python
img = ndimage.rotate(img,50,mode='constant')
plt.imshow(img,cmap=plt.cm.gray)
plt.title('Rotate the image')
plt.show()

sx = ndimage.sobel(img,axis=0,mode='constant')
sy = ndimage.sobel(img, axis=1, mode='constant')

plt.imshow(sx)
plt.axis('off')
plt.title('Sobel X')
plt.show()

plt.imshow(sy)
plt.axis('off')
plt.title('Sobel Y')
plt.show()
```
